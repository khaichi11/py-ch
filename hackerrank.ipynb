```python
if __name__ == '__main__':
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())
    hasil = []
    for i in range(0,x+1):
        for j in range(0,y+1):
            for k in range(0,z+1):
                if (i+j+k) == n:
                    continue
                else:    
                    hasil.append([i,j,k])
                
    print(hasil)
```

```python
records = []

if __name__ == '__main__':
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name,score])
    
    # ambil scores dari list records
    scores = []
    for name, score in records:
        scores.append(score)
    # hilangkan duplikat dengan mengubah menjadi set
    scores = set(scores)
    # jadiin list lagi supaya ada index
    scores = list(scores)
    # urutin list
    scores.sort()
    # ambil terkecil kedua
    terkecil_kedua = scores[1]
    
    nama = []
    for name,score in records:
        if score == terkecil_kedua :
            nama.append(name)
            
    nama.sort()
    for i in nama :
        print(i)
        
    

    
```
