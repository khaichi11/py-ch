### List Comprehension

```python
if __name__ == '__main__':
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())
    hasil = []
    for i in range(0,x+1):
        for j in range(0,y+1):
            for k in range(0,z+1):
                if (i+j+k) == n:
                    continue
                else:    
                    hasil.append([i,j,k])
                
    print(hasil)
```

### Find the Runner-Up Score!

```python
if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split()) # map untuk mengubah input string menjadi integer
    
    array = list(arr)
    terbesar = max(array)
    while terbesar in array:
        array.remove(terbesar)
    
    # print(array_baru)
    terbesar_kedua = max(array)
    print(terbesar_kedua)
```

### Nested List

```python
records = []

if __name__ == '__main__':
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name,score])
    
    # ambil scores dari list records
    scores = []
    for name, score in records:
        scores.append(score)
    # hilangkan duplikat dengan mengubah menjadi set
    scores = set(scores)
    # jadiin list lagi supaya ada index
    scores = list(scores)
    # urutin list
    scores.sort()
    # ambil terkecil kedua
    terkecil_kedua = scores[1]
    
    nama = []
    for name,score in records:
        if score == terkecil_kedua :
            nama.append(name)
            
    nama.sort()
    for i in nama :
        print(i)
        
    

    
```

### List

```python
if __name__ == '__main__':
    N = int(input())
    
    array = []
    for i in range(N):
        perintah = input().split()
        if perintah[0] == 'insert':
            array.insert(int(perintah[1]), int(perintah[2]))
        elif perintah[0] == 'print':
            print(array)
        elif perintah[0] == 'remove':
            array.remove(int(perintah[1]))
        elif perintah[0] == 'append':
            array.append(int(perintah[1]))
        elif perintah[0] == 'sort':
            array.sort()
        elif perintah[0] == 'pop':
            array.pop()
        elif perintah[0] == 'reverse':
            array.reverse()
        else:
            print("Pilihan Salah")
```
