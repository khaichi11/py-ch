### Import Library

```python
import pandas as pd
from sklearn.model_selection import KFold, cross_validate
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score

# Memuat dataset
data_heart = pd.read_csv('/content/drive/MyDrive/AITF/Self-Practice: Strategi Perancangan Model/heart.csv')
```

### Melihat Dimensi Awal Data

```python
print("Dimensi awal:", data_heart.shape)
```

### Melakukan Penghapusan Pada Data Duplikat

```python
data_heart_clean = data_heart.drop_duplicates()
print("Dimensi setelah pembersihan:", data_heart_clean.shape)
```

### Memisahkan Fitur (X) dan Target (y)

```python

X = data_heart_clean.drop('target', axis=1)
y = data_heart_clean['target']
```

### Inisialisasi Model SVM dalam Pipeline Menggunakan kernel 'rbf'

```python
model_svm = make_pipeline(StandardScaler(), SVC(kernel='rbf', random_state=42))
```

### Konfigurasi Cross Validation dengan 5 Fold

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

### Metrik Evaluasi

```python
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score, average='binary'),
    'recall': make_scorer(recall_score, average='binary'),
    'f1': make_scorer(f1_score, average='binary')
}

results = cross_validate(model_svm, X, y, cv=kf, scoring=scoring)
```

### Hasil Evaluasi

```python
print(f"Rata-rata Akurasi: {results['test_accuracy'].mean():.2f}")
print(f"Rata-rata Precision: {results['test_precision'].mean():.2f}")
print(f"Rata-rata Recall: {results['test_recall'].mean():.2f}")
print(f"Rata-rata F1: {results['test_f1'].mean():.2f}")
```
