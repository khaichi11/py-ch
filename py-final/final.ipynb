```python
text = "apple banana orange banana apple"
kata = text.split()

hasil = {}

for k in kata:
    if k in hasil:
        hasil[k] = hasil[k] + 1
    else:
        hasil[k] = 1

print(hasil)
```

```python
data = [
    ("Andi", "A"),
    ("Budi", "B"),
    ("Cici", "A"),
    ("Dodi", "C"),
    ("Eka", "B")
]
data_new = {}
for nama, kelompok in data:
    if kelompok not in data_new:
        data_new[kelompok]=[]
    data_new[kelompok].append(nama)        
    
print (data_new)
```
