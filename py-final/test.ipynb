Analisis Penjualan Produk Harian

Anda sedang membuat modul analisis sederhana untuk sebuah aplikasi pembelajaran Python yang mencatat data penjualan produk setiap hari.

ðŸ“Œ Tugas

Buatlah sebuah program Python yang:

Membaca sebuah bilangan bulat n yang menyatakan jumlah produk

Untuk setiap produk, membaca:

Nama produk (string)

Daftar jumlah penjualan harian berupa bilangan bulat yang dipisahkan spasi

Menyimpan data produk dalam struktur data yang sesuai

Menghitung:

Total penjualan setiap produk

Rata-rata penjualan setiap produk

Nama produk dengan total penjualan tertinggi

Daftar produk dengan rata-rata penjualan â‰¥ 20

Rata-rata seluruh penjualan dari semua produk

Menampilkan hasil sesuai format

```python

  

if __name__ == "__main__":
    n = int(input("Masukkan jumlah siswa"))
    total_penjualan_rata = {}
    semua_penjualan ={}
    produk_banyak = []
    x = 0
    for i in range(n):
        barang = input().strip()
        total = list(map(int, input().split()))
        jumlah = sum(total)
        rata = sum(total)/len(total)
        semua_penjualan[barang] = jumlah
        total_penjualan_rata[barang] = round(rata,2)
   
    for barang, rata_jumlah in total_penjualan_rata.items():
        if rata_jumlah >= 20:
            produk_banyak.append(barang)
    rata_penjualan = sum(semua_penjualan.values())/len(semua_penjualan)
                
    print(total_penjualan_rata)  
    print(semua_penjualan) 
    print(max(semua_penjualan, key=semua_penjualan.get))
    print(produk_banyak)
    print(rata_penjualan)
    
  
    
```
