## Langkah 1: Mendeskripsikan Data (Data Understanding)

### Import Library

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
```

### Melakukan Load Dataset

```python
cancer = load_breast_cancer()
df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
```

### Melakukan Cek Dimensi dan Informasi Data

```python
print("Dimensi dataset (baris, kolom):", df.shape)
print("\nLima baris pertama dataset:")
display(df.head())

print("\nInformasi tipe data dan null:")
print(df.info())
```

### Memeriksa Statistik Deskriptif

```python
print("\nDeskripsi statistik dataset:")
display(df.describe())
```

### Memeriksa Distribusi Target

```python
print("\nDistribusi kelas target (0 = Malignant, 1 = Benign):")
print(df['target'].value_counts())
```

## Langkah 2: Eksplorasi Data (Data Visualization)

### Import Library

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Histogram untuk melihat distribusi fitur

```python
features_to_plot = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']
df[features_to_plot].hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle('Distribusi Fitur Utama')
plt.show()
```

### Scatter Plot

```python
sns.pairplot(df[features_to_plot + ['target']], hue='target', markers=["o", "s"])
plt.show()
```

### Heatmap Korelasi

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title('Heatmap Korelasi Seluruh Fitur')
plt.show()
```

## Langkah 3: Verifikasi Kualitas Data

### Cek Missing Value

```python
print("Jumlah nilai kosong (missing value) per kolom:")
print(df.isnull().sum().sum())
```

### Cek dan Hapus Duplikasi

```python
duplicate_count = df.duplicated().sum()
print(f"\nJumlah data duplikat: {duplicate_count}")

if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print("Duplikasi telah dihapus.")
else:
    print("Tidak ada duplikasi data.")
```

### Deteksi dan Penanganan Outlier dengan IQR

```python
print("\nDeteksi Outlier menggunakan metode IQR:")
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = ((df < lower_bound) | (df > upper_bound)).sum()
print("Jumlah outlier per kolom:\n", outliers[outliers > 0])
```

## Langkah 4: Kriteria dan Teknik Pemilihan Data

### Import Library

```python
from sklearn.model_selection import train_test_split
```

### Memisahkan Fitur dan Target

```python
X = df.drop('target', axis=1)
y = df['target']
```

### Teknik Pemilihan Data (Split)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

### Verifikasi Distribusi

```python
print("Distribusi kelas data training:\n", y_train.value_counts(normalize=True))
print("\nDistribusi kelas data testing:\n", y_test.value_counts(normalize=True))
```

## Langkah 5: Menentukan Attributes dan Records

### Identifikasi Attributes (Kolom)

```python
attributes = X_train.columns.tolist()
print(f"Jumlah Attributes (Fitur): {len(attributes)}")
print(f"Daftar Attributes (5 pertama): {attributes[:5]}")
```

### Identifikasi Records (Baris)

```python
print(f"\nJumlah Records Data Training: {X_train.shape[0]}")
print(f"Jumlah Records Data Testing: {X_test.shape[0]}")
print(f"Total Records Keseluruhan: {X_train.shape[0] + X_test.shape[0]}")
```

## Langkah 6: Integrasi dan Formatting Data

### Import Library

```python
from sklearn.preprocessing import MinMaxScaler
```

###Integrasi (Penggabungan Training dan Testing untuk Formatting seragam)

```python
merged_df = pd.concat([X_train, X_test], axis=0)
```

### Formatting Nama Kolom (Mengganti spasi dengan underscore)

```python
merged_df.columns = [col.replace(' ', '_') for col in merged_df.columns]
print("Nama kolom setelah formatting (contoh):", merged_df.columns[:3].values)
```

### Normalisasi Data (Scaling)

```python
scaler = MinMaxScaler()
merged_df_scaled = pd.DataFrame(
    scaler.fit_transform(merged_df),
    columns=merged_df.columns
)
```

### Menampilkan hasil akhir data yang siap pakai

```python
print("\nDataset setelah Integrasi, Formatting, dan Normalisasi:")
display(merged_df_scaled.head())
```

# Memastikan kualitas integrasi (tidak ada null baru)

```python
print("\nCek missing value pasca-integrasi:", merged_df_scaled.isnull().sum().sum())
```
